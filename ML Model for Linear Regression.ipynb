```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df= pd.read_csv('homeprices.csv')
df
```

```python
plt.scatter(df.area,df.price)
```

```python
plt.scatter(df.area,df.price, color='Blue', marker='*')
```

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price, color='Blue', marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[34487]])
```

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price, color='Blue', marker='*')
plt.plot(df.area,reg.predict(df[['area']]),color='Green')
```

```python
new_df=pd.read_csv('areas.csv')
new_df
```

```python
prediction=reg.predict(new_df)
```

```python
new_df['PREDICTIONS']=prediction
```
